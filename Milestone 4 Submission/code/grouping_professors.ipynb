```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_csv('data_professors.csv')
```

```python
df = df.fillna(0)
```

```python
df.head(2)
```

```python
x = df.drop(columns=['professor'])
```

```python
model = KMeans(n_clusters=10, init='k-means++')
model.fit(x)
```

```python
labels = model.predict(x)
labels
```

```python
df['label'] = labels
```

```python
mca = prince.MCA()
mca = mca.fit(x) # same as calling ca.fs_r(1)
mca = mca.transform(x) # same as calling ca.fs_r_sup(df_new) for *another* test set.
print(mca)
```

```python
df['mca_0'] = np.array(mca[0])
df['mca_1'] = np.array(mca[1])
```
